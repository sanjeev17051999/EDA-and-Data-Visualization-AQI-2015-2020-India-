### Dealing with the missing values

```python
df.isna()
```

```python
df.isna().T
```

```python
df.isna().sum()
```

```python
# Finding null values with the help of HeatMap

plt.figure(figsize=(10,6))  # Adjust the width and height as per your preference

sns.heatmap(df.isna(), cmap="YlGnBu", cbar=True, mask=False)

plt.show()
```

```python
# Fill missing values with mean for float columns
float_columns = df.select_dtypes(include=['float']).columns
df[float_columns] = df[float_columns].fillna(df[float_columns].mean())

```

```python
# Delete rows with missing values in categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
df = df.dropna(subset=categorical_columns)

# Check the resulting DataFrame
print(df)
```

### Finding outliers for PM2.5

```python
# We will use IQR Method to find the Outliers
# First we calculate the percentile

df.describe()
```

```python
plt.figure(figsize=(14,12))

sns.distplot(df['PM2.5'])
```

The distribution is right-skewed because it’s longer on the right side of its peak

```python
# To see the outlier clearly
plt.figure(figsize=(10,8))
sns.boxplot(df['PM2.5'])
```

### Outliers

```python
# To find the limits
upper_limit = df['PM2.5'].mean()+ 3*df['PM2.5'].std()

lower_limit = df['PM2.5'].mean()- 3*df['PM2.5'].std()

upper_limit,lower_limit
```

```python
# Find the outliers
df.loc[(df['PM2.5']<lower_limit) | (df['PM2.5']>upper_limit)]
```

```python
# Triming = Deleting the outlier data
new_df = df.loc[(df['PM2.5']>lower_limit) & (df['PM2.5']<upper_limit)]

print('old data:',len(df))

print('new data:',len(new_df))

print('outliers',len(df)-len(new_df))
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(new_df['PM2.5'])
```

### Data visualization and insights - 

```python
# Line plot of PM2.5 over time
plt.figure(figsize=(16, 8))
sns.lineplot(data=new_df, x='Date', y='PM2.5')
plt.title('PM2.5 Levels over Time')
plt.xticks(rotation=45)
plt.show()

```

This plot shows the trend of PM2.5 levels over time. You can observe if there are any patterns or fluctuations in the PM2.5 levels throughout the recorded period.



```python
# Boxplot of AQI distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=new_df, y='AQI')
plt.title('Distribution of AQI')
plt.show()

```

The boxplot provides an overview of the distribution of AQI values. You can identify the median, quartiles, and any outliers in the data, helping you understand the overall air quality conditions.



```python
#  Bar plot of AQI Bucket counts
plt.figure(figsize=(8, 6))
sns.countplot(data=new_df, x='AQI_Bucket')
plt.title('Count of AQI Buckets')
plt.xticks(rotation=45)
plt.show()

```

This plot displays the count of different AQI buckets. It gives an idea of how frequently each air quality category occurs, helping identify the dominant AQI buckets in the dataset.



```python
#  Scatter plot of PM2.5 vs. PM10
plt.figure(figsize=(8, 6))
sns.scatterplot(data=new_df, x='PM2.5', y='PM10')
plt.title('PM2.5 vs. PM10')
plt.show()

```

By plotting PM2.5 against PM10, you can observe the relationship between these two pollutants. It helps determine if there is a correlation or pattern between PM2.5 and PM10 levels.


```python
#  Histogram of O3 levels
plt.figure(figsize=(8, 6))
sns.histplot(data=new_df, x='O3', bins=20)
plt.title('O3 Distribution')
plt.show()

```

The histogram provides a visual representation of the distribution of O3 levels. It allows you to understand the frequency and concentration range of O3 in the dataset.



```python
#  Pairwise correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(new_df.corr(), cmap='coolwarm', annot=True)
plt.title('Pairwise Correlation Heatmap')
plt.show()

```

This heatmap displays the pairwise correlation coefficients between different columns. It helps identify the strength and direction of relationships between variables, highlighting any significant correlations.




```python
#  Bar plot of mean AQI by City
plt.figure(figsize=(10, 6))
mean_aqi_by_city = new_df.groupby('City')['AQI'].mean().sort_values(ascending=False)
sns.barplot(x=mean_aqi_by_city.index, y=mean_aqi_by_city.values)
plt.title('Mean AQI by City')
plt.xticks(rotation=45)
plt.show()

```

 This plot compares the average AQI values across different cities. It helps identify cities with higher or lower air pollution levels on average, providing insights into regional air quality differences.


```python
#  Line plot of AQI over time by City
plt.figure(figsize=(18, 9))
sns.lineplot(data=new_df, x='Date', y='AQI', hue='City')
plt.title('AQI Trends over Time by City')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This plot visualizes the trends of AQI over time for each city separately. It allows you to compare the air quality patterns and variations between cities, identifying any city-specific trends or anomalies.

```python
#Stacked area plot of Pollutant Levels over Time:

plt.figure(figsize=(12, 8))
pollutant_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
stacked_data = new_df[pollutant_columns + ['Date']].melt('Date', var_name='Pollutant', value_name='Level')
sns.lineplot(data=stacked_data, x='Date', y='Level', hue='Pollutant', palette='viridis', linewidth=1.5)
plt.title('Pollutant Levels over Time')
plt.xlabel('Date')
plt.ylabel('Level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This stacked area plot visualizes the trends of various pollutant levels over time..It helps identify any concurrent trends or variations in pollutant levels, such as high levels of PM2.5 and PM10 occurring together or variations in O3 levels.
